### Handle Missing Values

```python
df.isnull().sum()
```

```python
missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100 
print(pd.concat([missing, missing_pct], keys=['Count', 'Percent%'], axis=1))
```

```python
numeric_col = df.select_dtypes(include='number').columns 
categorical_col = df.select_dtypes(include='object').columns 

num_imputer = SimpleImputer(strategy='mean')
cat_imputer = SimpleImputer(strategy='most_frequent')

df[numeric_col] = num_imputer.fit_transform(df[numeric_col])
df[categorical_col] = cat_imputer.fit_transform(df[categorical_col])

df[numeric_col] = df[numeric_col].round(1)
```

### Encoding 

```python

for col in categorical_col: 
    print(f" {col} : {df[col].unique()}")
```

```python
le_edu = LabelEncoder()
df["Education_Level"] = le_edu.fit_transform(df["Education_Level"])

le_loan = LabelEncoder()
df["Loan_Approved"] = le_loan.fit_transform(df["Loan_Approved"])

print("\nEducation_Level mapping:")
for i, cls in enumerate(le_edu.classes_):
    print(f"{i} → {cls}")

print("\nLoan_Approved mapping:")
for i, cls in enumerate(le_loan.classes_):
    print(f"{i} → {cls}")
```

```python
cols = ["Employment_Status", "Marital_Status", "Loan_Purpose", "Property_Area", "Gender", "Employer_Category"]

ohe = OneHotEncoder(drop="first", sparse_output=False, handle_unknown="ignore")

encoded = ohe.fit_transform(df[cols])

encoded_df = pd.DataFrame(encoded, columns=ohe.get_feature_names_out(cols), index=df.index)

df = pd.concat([df.drop(columns=cols), encoded_df], axis=1)
```
